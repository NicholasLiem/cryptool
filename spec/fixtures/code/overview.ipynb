```python
# import preprocessed data as dataframe
import pandas as pd
import numpy as np
import os

# Import data
shows = pd.read_json('final_data\shows.json',orient='index')
countries = pd.read_json('final_data\countries.json',orient='index')
companies = pd.read_json('final_data\companies.json',orient='index')
groups = pd.read_json('final_data\groups.json',orient='index')
idols = pd.read_json('final_data\idols.json',orient='index')
albums = pd.read_json('final_data\\albums.json',orient='index')
songs = pd.read_json('final_data\songs.json',orient='index')
concerts = pd.read_json('final_data\concerts.json',orient='index')
awards = pd.read_json('final_data\\awards.json',orient='index')

entities = {"shows" : shows,
            "countries": countries,
            "companies": companies,
            "groups": groups,
            "idols": idols,
            "albums": albums,
            "songs": songs,
            "concerts": concerts,
            "awards": awards}
```

```python
def displayEntityTitle(title : str):
  print("========================================")
  print(title.upper())
  print("========================================")
```

```python
for key, entity in entities.items():
  cols = entity.columns.tolist()
  displayEntityTitle(key)
  for col in cols:
    print(col, ":", entity[col].isna().sum())
```
